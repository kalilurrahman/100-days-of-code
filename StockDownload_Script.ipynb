```python
try:
  import yfinance as yf
except:
  !pip install yfinance
  import yfinance as yf


NVDA = yf.Ticker("NVDA")
print(NVDA)
# get stock info
NVDA_info=NVDA.info
# get historical market data
NVDA_hist=NVDA.history(period="max")
# show actions (dividends, splits)
NVDA_action=NVDA.actions
# show dividends
NVDA_divi=NVDA.dividends
# show splits
NVDA_splits=NVDA.splits
```

```python
print(NVDA_hist.head())
print(NVDA_hist.tail())
```

```python
import csv
#file_name = GIT_PATH + 'Micosoft_stock.csv' 
NVDA_hist.to_csv('./testfile/NVidia_stock_history.csv')
print(NVDA_info)
with open('./testfile/NVidia_stock_info.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in NVDA_info.items():
       writer.writerow([key, value])
NVDA_action.to_csv('./testfile/NVidia_stock_action.csv')
NVDA_divi.to_csv('./testfile/NVidia_stock_dividends.csv')
NVDA_splits.to_csv('./testfile/NVidia_stock_spilts.csv')
```
